```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Kaggle/Titanic/data/train.csv')
test = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Kaggle/Titanic/data/test.csv')
gender_submission = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Kaggle/Titanic/data/gender_submission.csv')
```

```python
# train, testと共通する処理をそれぞれ実行する必要をなくすために結合
data = pd.concat([train, test], sort=False)
```

```python
data
```

```python
data['Sex'].replace(['male', 'female'], [0, 1], inplace=True)
```

```python
data['Embarked'].fillna(('S'), inplace=True)
data['Embarked'] = data['Embarked'].map({'S': 0, 'C': 1, 'Q': 2}).astype(int)
```

```python
data['Fare'].fillna(np.mean(data['Fare']), inplace=True)
```

```python
age_avg = data['Age'].mean()
age_std = data['Age'].std()

data['Age'].fillna(np.random.randint(age_avg - age_std, age_avg + age_std), inplace=True)
```

```python
delete_columns = ['Name', 'PassengerId', 'SibSp', 'Parch', 'Ticket', 'Cabin']
data.drop(delete_columns, axis=1, inplace=True)
```

```python
# ここで、最初に結合したtrainとtestを分ける
train = data[:len(train)]
test = data[len(train):]
```

```python
y_train = train['Survived']
X_train = train.drop('Survived', axis=1)
X_test = test.drop('Survived', axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(penalty='l2', solver='sag', random_state=0)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
y_pred[:20]
```

```python
sub = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Kaggle/Titanic/data/gender_submission.csv')
sub['Survived'] = list(map(int, y_pred))
```
